## Map alias 'country names' to one distinct identifying name

```python
# Aim: use 'Country Name' as a unique ID
# Step 1: Identify inconsistencies in country names across all 3 files (Corruption, GDP & Population)
# Note: we'll use the 'Corruption' file as the base 
```

```python
def create_LkupTable(country_name_tuple):
    country_lkup={}
    # pop_country_name_df.to_dict()
    for country in country_name_tuple:
        country_lkup[country]=True

    return country_lkup
```

```python
corrup_country_name_df = tuple(corruption_df['country_name'].unique())
```

```python
# Step 1A: Identify country names in 'GDP' file, but not in corruption file
gdp_countries_tuple = tuple(gdp_df[gdp_df.columns[0]])
dic = create_LkupTable(gdp_countries_tuple)

for country in corrup_country_name_df:
    if country not in dic:
        print(country)  
```

```python
# Step 1B: Identify country names in 'population' file, but not in corruption file
pop_countries_tuple = tuple(population_df['Country Name'])
dic = create_LkupTable(pop_countries_tuple)

for country in corrup_country_name_df:
    if country not in dic:
        print(country) 
```

```python
# Step 3: grab only the years listed below that are common across all 3 files
gdp_clean_df = gdp_df[[gdp_df.columns[0], '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', 
                       '2012', '2013', '2014']]

pop_clean_df = population_df[['Country Name', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
                              '2012', '2013', '2014']]

corr_clean_df = corruption_df[['country_name', 'indicator_name', 'indicator_code', '2002', '2003', '2004', '2005', '2006',
           '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']]
```

```python
# Step 4: replace country names in pop & gdp df's with updated names
def replace_country_name(df, mapper):
    df_ret = df
    country_field_name = df_ret.columns[0]
    for country in mapper:
        print(mapper[country])
        df_ret.loc[df_ret[country_field_name] == country, country_field_name] = mapper[country]
    
    return df_ret

replace_country_name(gdp_clean_df, gdp_map)

replace_country_name(pop_clean_df, pop_map)

gdp_clean_df
```

```python
# Renaming some columns
gdp_clean_df = gdp_clean_df.rename(columns={gdp_df.columns[0]: 'country_name'})
pop_clean_df = pop_clean_df.rename(columns={'Country Name': 'country_name'})
```
